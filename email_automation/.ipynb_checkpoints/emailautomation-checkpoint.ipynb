```python
#Email Autoamtion Using Selenium -Python

import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd



email_ids = "******"
passwords = "******"

to_emails = "*******"

sent_emails = []
sent_email_records = []

driver = webdriver.Chrome()
driver.maximize_window()


    
driver.get('https://www.gmail.com')
time.sleep(10)
    
driver.refresh()
time.sleep(10)


# Login to the mail

driver.find_element(By.XPATH, '//input[@id="identifierId"]').send_keys(email_ids)
driver.find_element(By.XPATH, '//span[contains(text(),"Next")]').click()
time.sleep(10)

driver.find_element(By.XPATH, '//input[@type="password"]').send_keys(password)
driver.find_elements(By.XPATH,'//span[contains(text(),"Next")]').click()
time.sleep(10)


#Finding the compose and Composingthe mail

driver.find_element(By.XPATH, '//input@[@class="agP aFW"]').send_keys(to_emails)
time.sleep(10)
driver.find_element(By.XPATH, '//input@[@name="subjectbox"]').send_keys("Enter the subject here")
time.sleep(10)
driver.find_element(By.XPATH, '//div@[@aria-label="Message Body"]').send_keys("Body of the mail")


#Sending the mail

driver.find_element(By.XPATH, '//div@[@class="T-I J-J5-Ji aoO v7 T-I-atl L3 T-I-Zf-aw2"]').click()
time.sleep(10)

#Logout of Email
driver.get("https://accounts.google.com/Logout?h1=en")
time.sleep(10)
```
